```python
import numpy as np
import fpw

# Generate some test data
t = np.linspace(0, 1, 100)
y = np.sin(2 * np.pi * t)
ivar = np.ones_like(y)
freqs = np.linspace(0, 10, 10000)
N_bins = 10

# Call the function
result = fpw.py_runFPW(t, y, ivar, freqs, N_bins)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(freqs, result)
plt.show()
```
