```python
import numpy as np
import fpw
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/data/sams_datasets/tydy.csv', header=None)

t = df[0].values
y = df[1].values
dy = df[2].values

df = pd.read_csv('/data/sams_datasets/fgrid.csv', header=None)

fgrid = df[0].values

plt.figure()
plt.errorbar(t, y, yerr=dy, fmt='.', color="black")
plt.show()
```

```python
#Run FPW on this data
Nbins = 10
ivar = 1/dy**2
result = fpw.py_runFPW(y, ivar, fgrid, t, Nbins)
```

```python
plt.figure(figsize=(16, 6))
plt.plot(fgrid, result, color="black")
plt.show()
```
